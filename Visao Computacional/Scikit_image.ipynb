```python
from skimage import data,filters
from skimage.color import rgb2gray , rgb2hsv, label2rgb 
from skimage.segmentation import active_contour,chan_vese,slic, mark_boundaries, felzenszwalb
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_two_images(image1,image2,cmap="gray"):
    plt.figure(figsize=(15,15))
    plt.subplot(1,2,1)
    plt.imshow(image1)
    plt.subplot(1,2,2)
    plt.imshow(image2,cmap=cmap)
```

```python
def show_image(image,title,cmap="gray"):
    plt.title(title)
    plt.imshow(image,cmap=cmap)
```

```python
def binary_image(image_gray,thresholding):
    return (image_gray>thresholding)*1
```

```python
img = data.coffee()
img_gray = rgb2gray(img)

plot_two_images(img,img_gray)
```

```python
img_hsv = rgb2hsv(img)

plot_two_images(img,img_hsv,cmap=None)
```

```python
plt.figure(figsize=(15,15))
for i in range(10):
    binarized_gray = (img_gray > i*0.1)*1
    plt.subplot(5,2,i+1)
    plt.title("thresholding:>"+str(round(i*0.1,1)))
    plt.imshow(binarized_gray,cmap="gray")
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,15))
threshold = filters.threshold_otsu(img_gray)
otsu_binary = binary_image(img_gray,threshold)

plt.subplot(2,2,1)
show_image(otsu_binary,"thresholding:>"+str(threshold))

threshold_nib = filters.threshold_niblack(img_gray)
niblack_binary = binary_image(img_gray,threshold_nib)

plt.subplot(2,2,2)
show_image(niblack_binary,"thresholding niblack")

threshold_sauvola = filters.threshold_sauvola(img_gray)
plt.subplot(2,2,3)
show_image(threshold_sauvola,"sauvola thresholding")

sauvola_binary = binary_image(img_gray,threshold_sauvola)

plt.subplot(2,2,4)
show_image(sauvola_binary,"thresholding sauvola")

```

```python
astronaut = data.astronaut()
astro_gray = rgb2gray(astronaut)

astro_gray_noiseless = filters.gaussian(astro_gray,1)

x1 = 220 + 100*np.cos(np.linspace(0,2*np.pi,500))
x2 = 110 + 100*np.sin(np.linspace(0,2*np.pi,500))

snake = np.array([x1,x2]).T

astronaut_snake = active_contour(astro_gray_noiseless,snake)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.imshow(astro_gray_noiseless,cmap="gray")
ax.plot(astronaut_snake[:,0],astronaut_snake[:,1],'-b',lw=5)
ax.plot(snake[:,0],snake[:,1],"--r",lw=5)
```

```python
chanvese_gray_astronaut = chan_vese(astro_gray,max_iter=100,extended_output=True)
```

```python
fig,axes = plt.subplots(1,3,figsize=(10,10))

ax = axes.flatten()

ax[0].imshow(astro_gray,cmap="gray")
ax[0].set_title("astronaut_gray")

ax[1].imshow(chanvese_gray_astronaut[0],cmap="gray")
title = "Chan-vese segmantation: {} iterations".format(len(chanvese_gray_astronaut[2]))
ax[1].set_title("title")

ax[2].imshow(chanvese_gray_astronaut[1],cmap="gray")
ax[2].set_title("last level")
```

```python
plt.figure(figsize=(15,15))

astronaut_seg = slic(astronaut,n_segments=100,compactness=1)
plt.subplot(1,2,1)
plt.imshow(astronaut)
plt.subplot(1,2,2)
plt.imshow(mark_boundaries(astronaut, astronaut_seg))
```

```python
astronaut_segments = slic(astronaut,n_segments=50,compactness=10)
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
show_image(astronaut,"original",cmap=None)
plt.subplot(1,2,2)
show_image(label2rgb(astronaut_segments,astronaut,kind="avg"),"resultado",cmap=None)
```

```python
plt.figure(figsize=(15,15))
astronaut_segments = felzenszwalb(astronaut,scale=2,sigma=5,min_size=100)
plt.subplot(1,2,1)
plt.imshow(astronaut)
plt.subplot(1,2,2)
plt.imshow(mark_boundaries(astronaut,astronaut_segments))
```
